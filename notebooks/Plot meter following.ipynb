```python
import json

results = json.load(open('../eval_results/meter_eval_results_1_100.json'))
```

```python
true_meters = results['true_meters']
pred_meters = results['pred_meters']
labels = results['labels']
```

```python
from sklearn.metrics import confusion_matrix
cf_matrix = confusion_matrix(true_meters, pred_meters)
```

```python
cf_matrix
```

```python
import seaborn as sns
sns.heatmap(cf_matrix, annot=True, xticklabels=labels, yticklabels=labels)
```

```python
import numpy as np
print(sum(np.array(true_meters)==np.array(pred_meters))/len(true_meters))
```
